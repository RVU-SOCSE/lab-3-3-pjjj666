```python
#without using update()
ph={
    'pj':1234,
    'saiu':5698,
    'shoru':789
}
print(ph)
print("After updating")
ph["shoru"]=888
print(ph)
```

```python
#values any data type int, float, string , list, etc
ph={
    'pj':1234,
    'saiu':5698,
    'shoru':789
}
print(ph)
ph["shoru"]={888,999,777}      #list allowed
print(ph)
ph["saiu"]={"saiu-home":7145585,"saiu-office":4569885}  # dictionary

print(ph)

print(ph["saiu"]["saiu-home"])
```
